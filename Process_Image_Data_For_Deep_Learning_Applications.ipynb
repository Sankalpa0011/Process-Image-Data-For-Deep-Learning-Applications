# **Resizing An Image (PIL)**

### **Resizing The Image Using Pillow Library**

```python
from PIL import Image
```

```python
# load the image
img  = Image.open("/content/happy-little-havanese-puppy-sitting-grass-orange-dog-45886384.jpg")
```

```python
# resize the image
img_resized = img.resize((200,200))
```

```python
# save the resized image
img_resized.save("dog_image_resized.jpg")
```

```python
# display the resized image from numpy array
img_res = mpimg.imread("/content/dog_image_resized.jpg")
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

## **RGB To Grayscale Conversion (OpenCV)**

```python
import cv2
```

```python
img = cv2.imread("/content/happy-little-havanese-puppy-sitting-grass-orange-dog-45886384.jpg")
```

```python
type(img)
```

```python
print(img.shape)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```
